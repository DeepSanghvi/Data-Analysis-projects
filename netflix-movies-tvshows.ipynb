```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/kaggle/input/netflix-shows-and-movies/Cleaned_Netflix_Data.csv')
data
```

```python
data.info()
```

```python
data = data.sort_values(by = "Release_Year")
data
```

```python
data.isna().sum()
```

```python
data = data.drop("Age_Certification", axis=1)
data
```

```python
data.isna().sum()
```

```python
data.Genres.fillna("Genre Unavailable", inplace=True)
data.Production_Countries.fillna("Country Unavailable", inplace=True)
data.isna().sum()
```

```python
data['IMDB_Score'].fillna(data['IMDB_Score'].mode()[0], inplace=True)
data['IMDB_Popularity'].fillna(data['IMDB_Popularity'].mode()[0], inplace=True)
data.dropna(subset=['Title'], inplace=True)
data.isna().sum()
```

```python
data
```

```python
data.to_csv('Netflix_data_cleaned.csv', index=False)
```

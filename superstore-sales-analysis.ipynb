# **Imports and Libraries**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# **Read CSV file** 

```python
df = pd.read_csv("D:\Jupyter_PY\SuperStoreOrders.csv")
df
```

```python
df.info()
```

# **Convert date to proper date-time format**

```python
# convert the 'Date' column to datetime format
df['order_date']= pd.to_datetime(df['order_date'])
df['ship_date']= pd.to_datetime(df['ship_date'])
df.info()
```

# **Convert Sales(obj) to float datatype**

```python
df['sales'] = df['sales'].str.replace(',', '',regex=True)
df['sales'] = df['sales'].astype(float)
df.info()
```

# **Sort data according to order date**

```python
df = df.sort_values(by="order_date")
df
```

# **Remove rows with shipping date 
#    earlier than order date**

```python
df = df[df['ship_date'] >= df['order_date']]
df
```

# **Check for missing, duplicate values and 
#     drop rows/cols if necessary**

```python
df.isna().sum()
```

```python
duplicates_df = df.duplicated()
print(duplicates_df)
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
columns_drop = ["market", "discount", "order_priority", "segment"]
df = df.drop(columns_drop, axis=1)
df.head()
```

# **Added a column to get Margin Profit in %**

```python
# Add profit margin column
df['%profit_margin'] = (df.profit / df.sales) * 100
df.head()
```

**Save cleaned csv file**

```python
df.to_csv('Cleaned_Superstore_sales.csv', index = False)
```
